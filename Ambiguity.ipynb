```python
def get_ambiguity_data(person):
    '''This function gets words marked relevant and non-relevant by the labeller'''
    
    N = len(person[-1])
    
    ambiguous = set()
    for i in range(N):
        if person[-1][i][3] == 'a':
            ambiguous.add(person[-1][i][0])
            
    nonambiguous = set()
    for i in range(N):
        if person[-1][i][3] == 'b':
            nonambiguous.add(person[-1][i][0])
            
    return ambiguous, nonambiguous
```

```python
ambiguous_argota , nonambiguous_argota = get_ambiguity_data(argota)
ambiguous_cata , nonambiguous_cata = get_ambiguity_data(cata)
ambiguous_manuel, nonambiguous_manuel = get_ambiguity_data(manuel)
ambiguous_marta , nonambiguous_marta = get_ambiguity_data(marta)
ambiguous_shivangi , nonambiguous_shivangi = get_ambiguity_data(shivangi)
ambiguous_uxue , nonambiguous_uxue = get_ambiguity_data(uxue)
```

```python
def find_ambiguity(relevant, ambiguous, nonambiguous):
    ambig = ambiguous.intersection(relevant)
    nonambig = nonambiguous.intersection(relevant)
    
    return ambig, nonambig
```

```python
ambig_cata, nonambig_cata = find_ambiguity(relevant_cata, ambiguous_cata, nonambiguous_cata)
ambig_manuel, nonambig_manuel = find_ambiguity(relevant_manuel, ambiguous_manuel, nonambiguous_manuel)
ambig_marta, nonambig_marta = find_ambiguity(relevant_marta, ambiguous_marta, nonambiguous_marta)
ambig_shivangi, nonambig_shivangi = find_ambiguity(relevant_shivangi, ambiguous_shivangi, nonambiguous_shivangi)
ambig_uxue , nonambig_uxue  = find_ambiguity(relevant_uxue , ambiguous_uxue , nonambiguous_uxue )
```

```python
all_labels_ambiguity = [ambig_cata, nonambig_cata, ambig_manuel, nonambig_manuel, ambig_marta, nonambig_marta, ambig_shivangi, nonambig_shivangi, ambig_uxue , nonambig_uxue]
decision_nonambiguous, decision_ambiguous = deciding_relevance(3, all_labels_ambiguity, marta)
```
